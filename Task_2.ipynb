# Zadanie: pozostałe metryki
Policz pozostałe metryki dla tego modelu, zrób wizualizacyjną ocenę i następnie porównaj ostatni model z pierwszym, który stworzyliśmy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, PowerTransformer, StandardScaler, PolynomialFeatures
from sklearn.linear_model import ElasticNet, LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from math import sqrt

# 1. Załaduj dane
bike_data = pd.read_csv('daily-bike-share.csv')

# 2. Zdefiniuj cechy i target
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
categorical_features = ['season', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
target = 'rentals'

# 3. Prosty model liniowy (tylko temperatura)
X_simple = bike_data[['temp']].copy()
y_simple = bike_data[target].copy()
X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(
    X_simple, y_simple, test_size=0.2, shuffle=False, random_state=42)

lr_model = LinearRegression()
lr_model.fit(X_train_simple, y_train_simple)
y_predict_test = lr_model.predict(X_test_simple)

# 4. Zaawansowany model z preprocessingiem
X = bike_data[numeric_features + categorical_features].copy()
y = bike_data[target].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False, random_state=42)

# Definicja transformatorów
numeric_transformer = Pipeline(steps=[
    ('logtransformer', PowerTransformer()),
    ('standardscaler', StandardScaler()),
    ('polynomialfeatures', PolynomialFeatures())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Finalny Pipeline
final_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('regressor', ElasticNet())])

# Hiperparametry
params = {
    'preprocessor__num__polynomialfeatures__degree': [1, 2, 3],
    'regressor__alpha': [1e-3, 1e-1, 1.0, 10.0],
    'regressor__l1_ratio': [0.1, 0.5, 0.9]
}

# Inicjalizacja KFold (poprawione - bez random_state gdy shuffle=False)
cv = KFold(n_splits=3, shuffle=False)

# Grid Search
print("Rozpoczynam trenowanie zaawansowanego modelu...")
final_polynomial_regression_gridsearch = GridSearchCV(
    final_pipeline,
    params,
    scoring='neg_mean_squared_error',
    cv=cv,
    n_jobs=-1)

final_polynomial_regression_gridsearch.fit(X_train, y_train)
print("Trenowanie zakończone!")

# Pobranie najlepszego modelu
final_polynomial_regression_model = final_polynomial_regression_gridsearch.best_estimator_
print("\nNajlepsze hiperparametry:", final_polynomial_regression_gridsearch.best_params_)

# Predykcje i ocena modelu
y_pred_train_final = final_polynomial_regression_model.predict(X_train)
y_pred_test_final = final_polynomial_regression_model.predict(X_test)

# 5. Obliczenie metryk
print("\nMETRYKI DLA OSTATNIEGO MODELU:")
print(f'Train R^2: {r2_score(y_train, y_pred_train_final):.4f}')
print(f'Test R^2: {r2_score(y_test, y_pred_test_final):.4f}')
print(f'Train MAE: {mean_absolute_error(y_train, y_pred_train_final):.2f}')
print(f'Test MAE: {mean_absolute_error(y_test, y_pred_test_final):.2f}')
print(f'Train MAPE: {mean_absolute_percentage_error(y_train, y_pred_train_final):.4f}')
print(f'Test MAPE: {mean_absolute_percentage_error(y_test, y_pred_test_final):.4f}')
print(f'Train RMSE: {sqrt(mean_squared_error(y_train, y_pred_train_final)):.2f}')
print(f'Test RMSE: {sqrt(mean_squared_error(y_test, y_pred_test_final)):.2f}')

# 6. Wizualizacja
errors_final = y_pred_test_final - y_test

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(x=y_test, y=errors_final, alpha=0.25)
plt.axhline(0, color="r", linestyle="--")
plt.xlabel('Rzeczywista wartość')
plt.ylabel('Reszty')
plt.title('Wykres reszt - model finalny')

plt.subplot(1, 2, 2)
plt.hist(errors_final, bins=20)
plt.axvline(errors_final.mean(), color='k', linestyle='dashed', linewidth=1)
plt.title(f'Histogram reszt (średnia={np.round(errors_final.mean(), 2)})')

plt.tight_layout()
plt.show()

# 7. Porównanie modeli
print("\nPORÓWNANIE MODELI:")
print(f"Pierwszy model (prosta regresja liniowa - tylko temperatura):")
print(f"- Test R^2: {r2_score(y_test_simple, y_predict_test):.4f}")
print(f"- Test RMSE: {sqrt(mean_squared_error(y_test_simple, y_predict_test)):.2f}")

print(f"\nOstatni model (złożony z preprocessingiem):")
print(f"- Test R^2: {r2_score(y_test, y_pred_test_final):.4f}")
print(f"- Test RMSE: {sqrt(mean_squared_error(y_test, y_pred_test_final)):.2f}")

# Obliczenie poprawy
rmse_simple = sqrt(mean_squared_error(y_test_simple, y_predict_test))
rmse_final = sqrt(mean_squared_error(y_test, y_pred_test_final))
improvement = rmse_simple - rmse_final
improvement_percent = improvement / rmse_simple * 100
print(f"\nPoprawa RMSE: {improvement:.2f} ({improvement_percent:.1f}%)")
```

## Na podstawie przedstawionych wyników możemy wyciągnąć następujące wnioski:
### Znaczna poprawa jakości modelu:
- Złożony model z preprocessingiem osiągnął RMSE = 454.97 w porównaniu do 630.13 w prostym modelu, co oznacza 27.8% poprawę dokładności predykcji.
- Współczynnik determinacji R² wzrósł z 0.19 do 0.58, co wskazuje, że nowy model wyjaśnia znacznie większą część wariancji danych.
### Efektywność preprocessingu:
- Użycie wszystkich dostępnych cech (numerycznych i kategorycznych) zamiast tylko temperatury dało znaczącą poprawę
- Transformacje danych (standaryzacja, kodowanie kategoryczne) okazały się skuteczne
### Optymalna złożoność modelu:
- Hiperparametr degree=1 dla PolynomialFeatures wskazuje, że nieliniowość nie była konieczna - liniowe zależności wystarczyły
- Wartość alpha=0.1 dla ElasticNet sugeruje umiarkowaną regularyzację, co zapobiega przeuczeniu
### Praktyczne znaczenie poprawy:
- Redukcja błędu predykcji o ~175 wypożyczeń dziennie to znacząca wartość biznesowa
- Model może być znacznie bardziej użyteczny w praktycznych zastosowaniach
### Dalsze możliwości optymalizacji:
- Można przetestować inne metody preprocessingu (np. inne transformacje zmiennych)
- Warto rozważyć inne algorytmy (np. Random Forest, Gradient Boosting)
- Można dokładniej tunować hiperparametry w szerszych zakresach

Podsumowując, złożony model jest znacznie lepszy od prostego modelu liniowego, zarówno pod względem statystycznym (metryki), jak i praktycznym (dokładność predykcji). Wybór optymalnych hiperparametrów zapobiegł przeuczeniu, zachowując dobre właściwości uogólniające modelu.
